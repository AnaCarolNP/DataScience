
Faça uma função que recebe uma string e retorna a mesma string invertida.


```python
def str_invertida():
  invertida = ""
  palavra = input('Digite uma palavra: ')
  for i in range(len(palavra)-1, -1, -1):
    invertida += palavra[i]
  print(invertida)

str_invertida()
```
