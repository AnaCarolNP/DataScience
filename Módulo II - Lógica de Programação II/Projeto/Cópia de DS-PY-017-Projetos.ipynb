```python
# Exemplo de projeto - Análise de dados do Referendo da UE no Reino Unido

# Dados fictícios (pode substituir por dados reais)
referendum_data = {
    "Inglaterra": ("Leave", 51.9),
    "Escócia": ("Remain", 62.0),
    "País de Gales": ("Leave", 52.5),
    "Irlanda do Norte": ("Remain", 55.8)
}

# Função para calcular a média de votos do referendo
def calcular_media_votos(data):
    total_votos = len(data)
    soma_votos = sum([voto[1] for voto in data.values()])
    return soma_votos / total_votos

# Função para exibir os resultados do referendo
def exibir_resultados(data):
    for regiao, resultado in data.items():
        print(f"{regiao}: Votação - {resultado[0]}, Percentual - {resultado[1]}%")

# Lambda para verificar se o percentual de votos é maior ou igual a 50%
votos_majoritarios = lambda votos: votos[1] >= 50.0

# Manipulação de arquivos para salvar os resultados
try:
    with open("resultados_referendo.txt", "w") as arquivo:
        for regiao, resultado in referendum_data.items():
            arquivo.write(f"{regiao}: {resultado[0]}, {resultado[1]}%\n")
except IOError:
    print("Erro ao manipular o arquivo!")

# Chamada das funções
exibir_resultados(referendum_data)
media = calcular_media_votos(referendum_data)
print(f"Média de votos: {media}")
votos_ganhadores = dict(filter(votos_majoritarios, referendum_data.items()))
print("Regiões com votos majoritários:")
exibir_resultados(votos_ganhadores)
```
